```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = np.linspace(-10,10,100)
```

```python
m, c = 5,50
```

```python
noise = np.random.randint(-10,10,100)
```

```python
y= m*X+c+noise
```

```python
plt.style.use ("fivethirtyeight")
```

```python
plt.scatter(X, y)
plt.show()
```

```python
class CustomLinearRegression:
    def __init__ (self, lr=0.01, max_steps = 20):
        self.lr=lr
        self.max_steps= max_steps
        
    def fit (self, X, y):
        self.coef = np.random.random()
        
        for i in range (self.max_steps):
            self.gradient_descend
            
    def gradient_descend(self, X, y):
        delta_coef = self.gradient(X,y)
        self.coef+= delta_coef
        
    def gradient (self, X , y):
        yp= self.predict
        d2_e = np.mean((yp-y)*X)
        return -d2_e+lr
    
    def predict (self, X):
        return X*(self.coef)
        
        
```

```python
model = CustomLinearRegression(lr=0.0001,max_steps=2000)
```

```python
model.fit (X, y)
```
